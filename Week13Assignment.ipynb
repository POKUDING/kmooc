## 여러 파일의 내용을 합쳐서 wordcloud 만들기

본 과제에서는 여러분들이 여러 파일을 읽어 들여서 third-party package인 wordcloud를 이용하여 wordcloud를 만들 수 있는 지를 확인하는 과제를 수행할 것입니다. 8주차에서 여러분이 수행했던 과제와 마찬가지로, 다음에 주어진 코드와 설명을 잘 읽고 이해하면서 `__(1)___`에서 `__(4)__`에 들어갈 알맞은 답을 선택하기 바랍니다. 

루이스 캐롤(Lewis Carroll)의 이상한 나라의 앨리스(Alice's Adventures in Wonderland)와 토스토예프스키(Tostoevsky)의 '죄와 벌(crime and punishment)', 그리고 미국 헌법 전문(Constitution of the United Status)을 모두 읽어들여서 하나의 큰 텍스트로 묶은 후 wordcloud를 만들기 바랍니다.

1. wordcloud라는 파이썬 패키지에서 WordCloud라는 클래스를 불러들입니다.
2. glob을 이용하여 alice.txt, dostoevsky.txt, 그리고 constituion.txt를 차례로 열고, read()라는 메소드를 이용해 텍스트 내용을 text라는 변수에 할당합니다.
3. texts라는 빈 문자열을 하나 만든 후, 읽어 들인 text의 내용을 연결(concatenate)하여, texts라는 문자열이 모든 파일들의 내용을 포함할 수 있도록 만듭니다. 
3. WordCloud()라는 클래스의 인스턴스를 wc라는 변수에 할당합니다.
4. WordCloud라는 클래스에 있는 generate라는 메소드를 이용해 wordcloud 객체를 생성합니다.
5. matplotlib.pyplot을 plt를 이용하여 불러들입니다.
6. 생성된 wordcloud객체를 imshow를 이용하여 그림으로 만듭니다.
7. 때로는 불필요하게 자주 등장하는 단어들이 보일때가 있는데, worldcloud의 STOPWORDS(불용어)라는 클래스를 사용하는 방법을 이해하여 사용합니다. 

### 위의 1번부터 8번의 과정을 아래에 코드로 작성하였습니다. 

코드와 위의 설명을 보고 코드의 작성 흐름을 이해한 후, 다음 __(1)__에서 __(4)__까지 들어갈 명령어로 옳은 것을 고르시기 바랍니다. 문제를 풀기 위해서 아래의 셀과 같이 빈 코드 셀을 만든 후, 코드를 따라 적어 보면서 원하는 결과가 나타나는 지 확인하면 학습효과가 더 높을 것입니다.

```python
import glob
from wordcloud import WordCloud
```

```python
from wordcloud import STOPWORDS
stopwords = set(STOPWORDS)
# stopwords.add("know")
# one을 불용어 처리하기
# stopwords.add("one")

# alice.txt, dostoevky.txt, constitution.txt 만 선택하기
for file in glob.glob("*.txt"):
    texts = ''
    # 선택된 파일 이름 확인하기
    print(file)
    print("*"*50)
    f = open(file)
    text = f.read()
    # 각 파일의 첫 200글자 정도만 확인하기
    print(text[:200])
    print("-"*50)
    texts += text
    f.close()
    wc = WordCloud(stopwords=stopwords).generate(texts)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.imshow(wc)
plt.axis("off")
plt.show()
```
